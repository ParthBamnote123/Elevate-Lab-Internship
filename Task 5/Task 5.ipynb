Use sns.pairplot(), sns.heatmap() for visualization

```python
# sns.pairplot() [Survived And Age Plot ]
sns.pairplot(df[['Survived','Age']])
x = df['Survived']
y = df['Age']
plt.scatter(x,y)
plt.show()
```

```python
# sns.heatmap() [Survived And Age Plot ]
sns.heatmap(df[['Survived','Age']])
x = df['Survived']
y = df['Age']
plt.scatter(x,y)
plt.show()
```

Plot histograms, boxplots, scatterplots

```python
#Sex(Male,Female) with Age
plt.hist(df[df['Sex'] == 'male']['Age'],color='blue')
plt.hist(df[df['Sex'] == 'female']['Age'],color ='Pink')
x = df['Sex']
y = df['Age']
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
male_ages = df[df['Sex'] == 'male']['Age'].dropna()
female_ages = df[df['Sex'] == 'female']['Age'].dropna()

plt.boxplot([male_ages, female_ages], labels=['Male', 'Female'])
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Age Distribution by Sex')
plt.show()
```

```python
#Fare an Age
plt.scatter(df['Age'],df['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
