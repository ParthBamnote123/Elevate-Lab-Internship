```python
import sqlite3
```

```python
# Step 1: Connect to SQLite DB (creates one if it doesn't exist)
conn = sqlite3.connect("sales_data.db")


```

```python
# Step 2: Define  SQL query
query = "SELECT product, SUM(quantity) AS total_qty, SUM(quantity * price) AS revenue FROM sales GROUP BY product"
```

```python
# Step 3: Run query and load results into a pandas DataFrame
import pandas as pd
df = pd.read_sql_query(query, conn)
```

```python
# Step 4: Print the result
print(df)
```

```python
# Step 5: Close the connection
conn.close()
```

```python
import matplotlib.pyplot as plt
# Plot bar chart
df.plot(kind='bar', x='product', y='revenue')
# Customize plot
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")

# Show chart
plt.show()
```
